# Shadiyana Data Analysis - Complete Script-
## 1. Setup and Load Data


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("--- SCRIPT STARTED ---")
print("\nLoading data files...")
customer_df = pd.read_excel('/content/Customer information.xlsx')
vendor_df = pd.read_excel('/content/vendor information.xlsx')
print("Data loaded successfully.")
```

## 2. Exploratory Data Analysis (EDA)

```python
# Customer information overview
print("\n Customer Data Overview:")
print(customer_df.head())
print("\nCustomer Data Shape:", customer_df.shape)
print(customer_df.info())
# Vendour information
print("\nVendor Data Overview:")
print(vendor_df.head())
print("\nVendor Data Shape:", vendor_df.shape)
print(vendor_df.info())
```

##3. Clean and Prepare Data

```python
print("\nCleaning and preparing data...")

# Clean column names to remove leading/trailing spaces
customer_df.columns = customer_df.columns.str.strip()
vendor_df.columns = vendor_df.columns.str.strip()
print("- Column names cleaned.")
# Handle Missing Values in 'Number of Guests' by filling with the mean
mean_guests = customer_df['Number of Guests'].mean()
customer_df['Number of Guests'].fillna(mean_guests, inplace=True)
print(f"- Missing guest values filled with average ({int(mean_guests)}).")
# Convert 'Event Date' to datetime format
customer_df['Event Date'] = pd.to_datetime(customer_df['Event Date'], errors='coerce')
customer_df.dropna(subset=['Event Date'], inplace=True)
print("- 'Event Date' column converted.")
# Create 'event_month' column for seasonal analysis
customer_df['event_month'] = customer_df['Event Date'].dt.month_name()
# Merge the dataframes for deeper analysis
merged_df = pd.merge(customer_df, vendor_df, on='Business Name', how='left')
merged_df.dropna(subset=['Budget Category', 'Sub Area'], inplace=True)
print("- Dataframes merged.")
print("Data cleaning and preparation complete.\n\n")
```

##  4. Initial Analysis & Plots

```python
print("\n Generating initial analysis plots...")
sns.set_style("whitegrid")

# Plot 1: Seasonal Demand
plt.figure(figsize=(12, 6))
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
sns.countplot(y=customer_df['event_month'], order=month_order, palette='viridis')
plt.title('Seasonal Demand: Number of Queries per Month', fontsize=16)
plt.xlabel('Number of Queries'); plt.ylabel('Month')
plt.savefig('1_seasonal_demand.png')
print("- Saved plot: 1_seasonal_demand.png")

# Plot 2: Guest Count Distribution
plt.figure(figsize=(12, 6))
sns.histplot(customer_df['Number of Guests'], bins=30, kde=True)
plt.title('Distribution of Guest Counts', fontsize=16)
plt.xlabel('Number of Guests'); plt.ylabel('Frequency')
plt.savefig('2_guest_count_distribution.png')
print("- Saved plot: 2_guest_count_distribution.png")
```

## 5. Deeper Analysis & Plots

```python
print(" Generating deeper analysis plots...")

# Plot 3: Area-Wise Vendor Popularity (Supply)
plt.figure(figsize=(12, 8))
sns.countplot(y=vendor_df['Sub Area'], order=vendor_df['Sub Area'].value_counts().index, palette='plasma')
plt.title('Vendor Distribution by Sub-Area', fontsize=16)
plt.xlabel('Number of Vendors'); plt.ylabel('Sub-Area')
plt.tight_layout(); plt.savefig('3_area_popularity.png')
print("- Saved plot: 3_area_popularity.png")

# Plot 4: Vendor Budget Category Distribution (Supply)
plt.figure(figsize=(10, 6))
vendor_df['Budget Category'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title('Distribution of Vendors by Budget Category', fontsize=16); plt.ylabel('')
plt.savefig('4_budget_category_distribution.png')
print("- Saved plot: 4_budget_category_distribution.png")

 # Plot 5: Customer Query Volume by Budget Category (Demand)
plt.figure(figsize=(12, 6))
sns.countplot(y=merged_df['Budget Category'], order=merged_df['Budget Category'].value_counts().index, palette='magma')
plt.title('Customer Query Volume by Vendor Budget Category', fontsize=16)
plt.xlabel('Number of Customer Queries'); plt.ylabel('Budget Category')
plt.savefig('5_customer_budget_preference.png')
print("- Saved plot: 5_customer_budget_preference.png")
```

## 6. Advanced Analysis & Plots

```python
print("\nGenerating advanced analysis plots...")

# Plot 6: Budget Mix within Top 10 Sub-Areas
top_10_areas = vendor_df['Sub Area'].value_counts().nlargest(10).index
top_areas_df = vendor_df[vendor_df['Sub Area'].isin(top_10_areas)]
plt.figure(figsize=(14, 8))
sns.countplot(y='Sub Area', hue='Budget Category', data=top_areas_df, order=top_10_areas, palette='coolwarm')
plt.title('Budget Category Mix in Top 10 Vendor Sub-Areas', fontsize=16)
plt.xlabel('Number of Vendors'); plt.ylabel('Sub-Area'); plt.legend(title='Budget Category')
plt.tight_layout(); plt.savefig('6_area_budget_mix.png')
print("- Saved plot: 6_area_budget_mix.png")

# Plot 7: Guest Count vs. Budget Category
plt.figure(figsize=(12, 7))
sns.boxplot(x='Budget Category', y='Number of Guests', data=merged_df, palette='pastel')
plt.title('Guest Count Distribution by Budget Category', fontsize=16)
plt.xlabel('Budget Category'); plt.ylabel('Number of Guests'); plt.ylim(0, 800)
plt.savefig('7_guest_vs_budget.png')
print("- Saved plot: 7_guest_vs_budget.png")

# Plot 8: Market Gap Analysis (Demand vs. Supply by Area)
supply_df = vendor_df.groupby('Sub Area').size().reset_index(name='vendor_count')
demand_df = merged_df.groupby('Sub Area').size().reset_index(name='query_count')
market_gap_df = pd.merge(supply_df, demand_df, on='Sub Area', how='left').fillna(0)
market_gap_df['demand_ratio'] = market_gap_df['query_count'] / (market_gap_df['vendor_count'] + 1)
market_gap_df_sorted = market_gap_df.sort_values(by='demand_ratio', ascending=False).head(15)
plt.figure(figsize=(14, 8))
sns.barplot(y='Sub Area', x='demand_ratio', data=market_gap_df_sorted, palette='rocket')
plt.title('Market Gap Analysis: Customer Demand Ratio by Sub-Area', fontsize=16)
plt.xlabel('Demand Ratio (Queries per Vendor)'); plt.ylabel('Sub-Area')
plt.tight_layout(); plt.savefig('8_market_gap_ratio.png')
print("- Saved plot: 8_market_gap_ratio.png")
```

##  7. Final Summary Tables

```python
print("\n[STEP 6/6] Generating final summary CSV files...")

# Table 1: Top 10 Most Queried Venues
top_venues = customer_df['Business Name'].value_counts().nlargest(10).reset_index()
top_venues.columns = ['Venue Name', 'Number of Queries']
top_venues.to_csv('9_top_10_queried_venues.csv', index=False)
print("- Saved table: 9_top_10_queried_venues.csv")

# Table 2: Customer Persona by Budget
persona_summary = merged_df.groupby('Budget Category')['Number of Guests'].agg(['mean', 'median', 'min', 'max', 'count']).round(0)
persona_summary.to_csv('10_customer_persona_summary.csv')
print("- Saved table: 10_customer_persona_summary.csv")

# Table 3: Full Market Gap Data
market_gap_df_full = market_gap_df.sort_values(by='demand_ratio', ascending=False)
market_gap_df_full.to_csv('11_market_gap_summary.csv', index=False)
print("- Saved table: 11_market_gap_summary.csv")

print("\n--- SCRIPT FINISHED ---")
```
